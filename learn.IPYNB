## 1. 三个工具分别做什么（用大白话）

- Git：本地的"版本管理器"，像游戏存档系统，能回到任何历史版本。
- GitHub：云端代码仓库，像网盘 + 协作平台，用来备份和团队协作。
- VS Code：代码编辑器，写代码和管理 Git 的主界面。

一句话理解：
你在 VS Code 写代码 -> 用 Git 记录改动 -> 推送到 GitHub 做云端备份与协作。

## 2. 先认识 8 个关键词（必须懂）

- 仓库（Repository / repo）：一个项目的版本库。
- 提交（commit）：一次保存记录。
- 暂存区（stage）：提交前的待保存清单。
- 分支（branch）：并行开发线，不影响主线。
- 合并（merge）：把分支改动合到主分支。
- 远程仓库（remote）：GitHub 上的仓库地址。
- 推送（push）：把本地提交上传到 GitHub。
- 拉取（pull）：把 GitHub 最新代码同步到本地。

## 8. 认证方式怎么选（新手建议）

你推送到 GitHub 需要认证，常见两种：

- HTTPS + Personal Access Token（PAT）
  - 优点：概念简单
  - 注意：GitHub 不再支持账号密码直接推送，要用 Token

- SSH Key
  - 优点：长期最省心，不用反复输 Token
  - 适合：会长期用 Git/GitHub 的同学

课堂建议：
- 第一次体验可用 HTTPS + Token
- 长期开发建议尽快切换 SSH

## 10. 每条核心命令是什么意思

- `git init`：让当前文件夹成为 Git 仓库。
- `git status`：查看哪些文件改了、是否已暂存。
- `git add .`：把改动加入暂存区（准备提交）。
- `git commit -m "..."`：形成一个历史快照（提交记录）。
- `git remote add origin URL`：关联 GitHub 仓库地址。
- `git push -u origin main`：把本地 main 分支上传到远程，并建立跟踪关系。

## 13. 常见报错与处理（课堂高频）

1. `git: command not found` / `git 不是内部或外部命令`
   - 原因：Git 未安装或 PATH 未生效
   - 处理：重装 Git，安装后重启终端

2. `Authentication failed`
   - 原因：HTTPS 使用了账号密码而不是 Token
   - 处理：使用 Personal Access Token 或改用 SSH

3. `rejected (non-fast-forward)`
   - 原因：远程比本地更新
   - 处理：
   ```bash
   git pull --rebase origin main
   git push
   ```

4. `fatal: remote origin already exists`
   - 原因：已添加过远程 origin
   - 处理：
   ```bash
   git remote set-url origin <new_url>
   ```

## 21. GitHub 上提 PR 的详细步骤（有权限也推荐走 PR）

1. 打开仓库页面：`https://github.com/loryemael/IMMC`
2. GitHub 会提示 `Compare & pull request`，点击进入

## 23. 本仓库高频问题与处理（非常实用）

### 情况 A：我不小心在 main 上改了代码
处理方式：把当前改动转到新分支再提交。
```bash
git checkout -b feat/rescue-work
git add .
git commit -m "feat: move work from main to feature branch"
git push -u origin feat/rescue-work
```

### 情况 B：push 时提示 non-fast-forward
说明远程比你新，先同步再推。
```bash
git pull --rebase origin main
git push
```

### 情况 C：出现冲突（conflict）
处理顺序：
1. 打开冲突文件，保留正确内容并删除冲突标记
2. `git add <冲突文件>`
3. 若在 rebase：`git rebase --continue`；若在 merge：`git commit`
4. 再 `git push`
