```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

# 1. 读取数据
file_name = '数据预处理COVID19 (不同国家).xlsx'
cols = ['country', 'code', 'date', 'total_deaths_per_million', 'gdp_per_capita', 'life_expectancy']
df_raw = pd.read_excel(file_name, usecols=cols, engine='openpyxl')
df_raw['date'] = pd.to_datetime(df_raw['date'])
df = df_raw.sort_values(['country', 'date']).drop_duplicates('country', keep='last').copy()

# 2. 定义三张图的配置（统一：越深越高）
configs = {
    "Medical_Quality": {
        "z": df['life_expectancy'],
        "colorscale": [[0, '#f7fbff'], [0.5, '#9ecae1'], [1, '#08306b']],
        "zmin": 60, "zmax": 82,
        "title": "Global Healthcare Quality (Life Expectancy)"
    },
    "Death_Rate": {
        "z": df['total_deaths_per_million'] / 10,
        "colorscale": [[0, '#fff5f0'], [0.5, '#fb6a4a'], [1, '#67000d']],
        "zmin": 0, "zmax": 350,
        "title": "COVID-19 Death Rate per 100k"
    },
    "GDP_per_Capita": {
        "z": df['gdp_per_capita'],
        "colorscale": [[0, '#f7fcf5'], [0.5, '#74c476'], [1, '#00441b']],
        "zmin": 0, "zmax": 65000,
        "title": "GDP per Capita (USD)"
    }
}

# 3. 循环生成并保存
for name, cfg in configs.items():
    fig = go.Figure(go.Choropleth(
        locations=df['code'],
        z=cfg['z'],
        text=df['country'],
        colorscale=cfg['colorscale'],
        zmin=cfg['zmin'],
        zmax=cfg['zmax'],
        marker_line_color='white',
        marker_line_width=0.5,
        colorbar=dict(title=name.replace("_", " "))
    ))
    
    fig.update_layout(
        title=dict(text=f"<b>{cfg['title']}</b>", x=0.5, y=0.9),
        height=600, width=1200,
        geo=dict(projection_type='natural earth', showframe=False),
        margin=dict(l=10, r=10, t=80, b=20)
    )
    
    # 保存为 PDF
    # 如果还是报错，尝试去掉 engine="kaleido"
    fig.write_image(f"{name}.pdf", engine="kaleido")
    print(f"已生成: {name}.pdf")
```
